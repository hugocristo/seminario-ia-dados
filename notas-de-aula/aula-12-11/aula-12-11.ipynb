```r
numeros <- c(10, 15, 20, 15, 20, 10, 5, 15, 20)
print(numeros)
print(numeros[5])
print(paste(numeros[5],numeros[7],"Hugo"))
# exemplo de uso em relatório de pesquisa
mean(numeros)
print(paste("A média das idades dos participantes foi",mean(numeros),"e a mediana foi",median(numeros)))
```

Quais níveis a coluna `vinculo`possui?

```r
# lista níveis únicos do vetor
print(unique(dados$vinculo))
# identifica linhas que apresentam o nível "docente"
substituir <- which(dados$vinculo == "docente")
print(substituir)
```

A substituição varre o vetor `substituir` e utiliza seus valores como as linhas da base de dados para saber quais linhas do vetor `vinculo` precisam ser alteradas.

```r
for (item in substituir) {
  print(item)
  # exibe qual é a linha
  print(dados$vinculo[item])
  # altera
  dados$vinculo[item] <- "professor"
  # exibe alteração
  print(dados$vinculo[item])
}
```
